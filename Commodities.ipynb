## Commodities: Atomoação web para atualização dos valores

Neste trabalho iremos buscar na internet, de forma automática, a cotação de diversas commodities, tais como soja, milho, trigo e petróleo.

A ideia é conferir se o valor está abaixo do preço ideal de compra imposto por uma empresa fictícia. Caso esteja, iremos marcar como uma opção de compra para a equipe de operações.

Usaremos o selenium, uma biblioteca que permite controlar o funcionamento de uma página e a interação com ela de forma automática. 

A base de dados foi adaptada de um dos cursos da Hashtag Programação.

```python
#!pip install selenium
```

```python
from selenium import webdriver
import pandas as pd
```

```python
# Abrindo o nabegador

navegador = webdriver.Chrome()
navegador.get("https://www.google.com")
```

```python
# Importando a base de dados

tabela = pd.read_excel("commodities.xlsx")

display(tabela)
```

```python
# Buscando as cotações

for linha in tabela.index:
    
    #Entrando no site que contém as cotações
    link = "https://www.melhorcambio.com/"+tabela.loc[linha, "Produto"]+"-hoje"
    link = link.replace("ó", "o").replace("ã", "a").replace("ç", "c").replace("ú","u").replace("é", "e").replace("á", "a")
    print(link)
    navegador.get(link)
        
    # Pegando e transformando a cotação em valor para o Python
    cotacao = navegador.find_element('xpath', '//*[@id="comercial"]').get_attribute('value') #outros atributos: .send_key("nome"), .click()
    cotacao = cotacao.replace(".", "").replace(",", ".")
    cotacao = float(cotacao)
    print(cotacao)
    
    # Preenchendo a coluna Preço Atual
    tabela.loc[linha, "Preço Atual"] = cotacao
```

```python
display(tabela)
```

```python
# Preenchendo a coluna Comprar

for linha in tabela.index:
    if tabela.loc[linha, "Preço Ideal"] >= tabela.loc[linha, "Preço Atual"]:
        tabela.loc[linha, "Comprar"] = "Sim"
    else:
        tabela.loc[linha, "Comprar"] = "Não"
```
